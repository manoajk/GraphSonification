```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
stores = pd.read_csv('stores.csv')
features = pd.read_csv('features.csv')
train = pd.read_csv('train.csv')
train.Weekly_Sales = train.Weekly_Sales.round(2)
store = 14 # has max standard deviation
mainDataset = features.loc[2366:2508] # Data of interest, has revenue data 
weeks = mainDataset.Date
numWeeks = len(weeks)
numDepts = 99
```

```python
revenuePerWeek = []
for week in weeks:
    revenuePerWeek.append(train[(train.Date == week) & (train.Store == 14)].Weekly_Sales.sum())
mainDataset['Revenue'] = revenuePerWeek
print(mainDataset)
```

mainDataset.to_csv('walmart_data.csv')
